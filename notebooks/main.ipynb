```python
#Import dependencies
import pandas as pd
import numpy as np
import os
from pdfquery import PDFQuery
import re
```

```python
def load_pdf(cv_path):
    cv_list=[]
    try:
        print(f"Full path: {cv_path}")
        for root, _, filenames in os.walk(cv_path):
            print(f"Reading files in {root}")
            for filename in filenames:
                if filename.endswith(('.pdf')):
                    file_path = os.path.join(root, filename)
                    cv_list.append(file_path)
        return cv_list
    except FileNotFoundError:
        print(f"The directory {cv_path} does not exist.")

cv_list=load_pdf("/content/drive/MyDrive/Pentakill/Resume")
```

```python
def extract_pdf_text(pdf_file):
    # Load the PDF file
    pdf = PDFQuery(pdf_file)
    pdf.load()

    # Use CSS-like selectors to locate the text elements
    text_elements = pdf.pq('LTTextLineHorizontal')
    # Extract the text from the elements and return as a list
    text = [re.sub(r'\s+', ' ', t.text) for t in text_elements]

    return text

cv_pdf = [extract_pdf_text(cv) for cv in cv_list]
```
